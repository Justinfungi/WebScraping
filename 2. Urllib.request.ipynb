# Parse.quote / urlcode

```python
headers={
    'User-Agent':"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}

name = parse.quote("周杰伦")
print(name)

url = "https://www.baidu.com/s?wd="
url = url+ name
print(url)

r = request.Request(url=url, headers = headers) # need to specify parameters
print(r)
response = request.urlopen(r)
print(response)
content = response.read().decode('utf-8')
print(content)
```

```python
headers={
    'User-Agent':"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}

data={
    'wd':'大张超市',
    'location': "中国河南省"
}
a= parse.urlencode(data)

url = "https://www.baidu.com/s?"
url = url+ a
print(url)

r = request.Request(url=url, headers = headers) # need to specify parameters
print(r)
response = request.urlopen(r)
print(response)
content = response.read().decode('utf-8')
print(content)
```

# Urllib ajax / movie list

```python
url= "https://movie.douban.com/j/chart/top_list?type=24&interval_id=100%3A90&action=&start=0&limit=20"

headers={
    'User-Agent':"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}

#(1) target
r = request.Request(url=url, headers = headers) # need to specify parameters
print(r)

#(2) request respnse
response = request.urlopen(r)
print(response)

content = response.read().decode('utf-8')
print(content)

#(3) download, open default is in gbk
with open("douban.js",'w', encoding='utf-8') as fp:
    fp.write(content)
```

```python
def create_request(page,header):
    base=url= "https://movie.douban.com/j/chart/top_list?type=24&interval_id=100%3A90&action=&"

    
    data={
        'start':(page-1)*20,
        'limit':20
    }
    
    data= parse.urlencode(data)
    
    url=base+data
    
    r = request.Request(url=url,headers=header)
    return r
    
def get_content(r):
    response = request.urlopen(r)
    content = response.read().decode('utf-8')
    return content
    
def download(page, content):
    with open("Source_Html/douban_"+str(page)+".js",'w', encoding='utf-8') as fp:
        fp.write(content)
    
if __name__ == '__main__':
    start_page = int(input("starting page"))
    end_page = int(input("ending page"))
    
    headers={
        'User-Agent':"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}

    
    for i in range(start_page,end_page+1):
        r = create_request(i,headers)
        c = get_content(r)
        download(i, c)
```

# Photos

```python
from urllib import request, parse
import requests
from lxml import etree

def create_request(page,header):
    base= "https://sc.chinaz.com/tupian/meinvxiezhen.html"

    if page==1 :
        url =base
    else:
        url = base[:-5]+'_'+str(page)+".html"

    #print(url)
    r = request.Request(url=url,headers=header)
    #print(r)
    return r

def get_content(r):
    response = request.urlopen(r)
    content = response.read().decode('utf-8')
    #print(content)
    return content

def download(page, content):
    tree = etree.HTML(content)
    name_list =tree.xpath('//div//div/img/@alt')
    img_list =tree.xpath('//div//div/img/@data-original')
    #img_list =tree.xpath('//div[@class="container"]//div[@class="item masonry-brick"]/img/@data-original') 
    print(len(name_list),len(img_list))
    # website will occur lazy load -> load less img for reduce lagging of website
    # turn src to src2, find the jpg one
    
    for i in range(len(name_list)):
        name=name_list[i]
        src=img_list[i]
        url = "https:"+src[:-6]+".jpg"
        #print(url)
        
        request.urlretrieve(url, filename ="img/"+name+".jpg")
        
        
    

    

    with open("Source_Html/douban_"+str(page)+".js",'w', encoding='utf-8') as fp:
        fp.write(content)


if __name__ == '__main__':
    start_page = int(input("starting page"))
    end_page = int(input("ending page"))
    
    headers={
        'User-Agent':"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}

    
    for i in range(start_page,end_page+1):
        r = create_request(i,headers)
        c = get_content(r)
        download(i, c)
```
