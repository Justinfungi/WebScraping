# lib and import

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# load 
r= requests.get("https://keithgalli.github.io/web-scraping/webpage.html")

webpage = bs(r.content)
webpage.prettify
```

# Module-1 Scraping links

```python
links =webpage.select("ul.socials a")
print(links)
links_actual = [link["href"] for link in links]
links_actual
```

```python
links = webpage.find("ul",attrs={"class":"socials"})
print(links)
links = links.find_all("a")
links_actual = [link["href"] for link in links]
links_actual
```

```python
links = webpage.select("li.social a")
print(links)
actual_links = [link['href'] for link in links]
actual_links
```

# Module-2 Scraping table

```python
import pandas as pd
table = webpage.select("table.hockey-stats")[0]
table
```

```python
col= table.find("thead").find_all("th")
print(col)
colname = [c.string for c in col]
colname
```

```python
table_rows = table.find("tbody").find_all("tr")
print(table_rows)
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)
df = pd.DataFrame(l, columns=colname)
df.head()
```

# Module-4 render pages

```python
files = webpage.select("div.block a")
relative_files = [f['href'] for f in files]


url = "https://keithgalli.github.io/web-scraping/"
for f in relative_files:
  full_url = url + f
  page = requests.get(full_url)
  bs_page = bs(page.content)
  secret_word_element = bs_page.find("p", attrs={"id": "secret-word"})
  secret_word = secret_word_element.string
  print(secret_word)
```
