# Request

```python
import requests

r = \
requests.get ("https://en.wikipedia.org/wiki/Natural_language_processing",
timeout =3)

print(r.raise_for_status)
print(r.status_code)
print(r.text)
print(r.text.encode("utf-8"))
```

```python
print(r.json) # for json data 
```

```python
with open('data−wikipedia−NLP.html', mode= 'wb' ) as fd :
    fd.write(r.content)
```

# BS

```python
from bs4 import BeautifulSoup
import re
import requests
r = \
requests.get ("https://en.wikipedia.org/wiki/Natural_language_processing",
timeout =3)

s = BeautifulSoup(r.text, "lxml")
print(s.prettify())
```

```python
print (s.get_text())
```

```python
s.p # get the first paragraph
```

```python
s.find(id = "footer")
```
