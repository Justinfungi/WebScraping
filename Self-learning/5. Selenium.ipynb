# Library

```python
import os
from selenium import webdriver
import time
```

```python
# get started of browser
driver= webdriver.Chrome('/home/fish/Documents/WebScraping/driver/chromedriver_ubuntu')
website = "https://www.jd.com/?country=CN"
driver.get(website)
```

# Content

```python
content=driver.page_source
print(content)
```

# Location of elements

```python
button = driver.find_element_by_class_name('button')
print(button)
```

```python
search_box = driver.find_element_by_id('key')
print(search_box)
```

# info of elements

```python
print(search_box.get_attribute('class'))
print(search_box.tag_name)
print(search_box.text)
```

# Interaction

```python
search_box.send_keys('camera')
time.sleep(2)

button.click()
time.sleep(2)

js_bottom = 'document.documentElement.scrollTop=100000'
driver.execute_script(js_bottom)
time.sleep(2)

#js_bottom = 'document.documentElement.scrollTop=2000'
#driver.execute_script(js_bottom)
#time.sleep(2)

next = driver.find_element_by_xpath('//a[@title="使用方向键右键也可翻到下一页哦！"]')
next.click()
time.sleep(2)

driver.back()
time.sleep(2)
driver.forward()
time.sleep(2)
```

## Modulized

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

def share_browser():
    chrome_options = Options()
    chrome_options.headless=True
    executable= "/home/fish/Documents/WebScraping/driver/chromedriver_ubuntu"
    driver = webdriver.Chrome(executable_path=executable, options=chrome_options)
    return driver

browser = share_browser()
url="https://www.jd.com/?country=USA"
browser.get(url)
browser.save_screenshot("aa.png")
```
