```python
import time
result = [[0,0,0,0,0]]
# content is text, time, comment, retweet, like
length = 0
while result[-1][0] != "Feb 20, 2022":
    #content = driver.find_elements(By.XPATH,'//*[@id]/span')
    ids = driver.find_elements(By.XPATH,'//*[@id]/div[2]/div/div[3]/a/time')
    comments = driver.find_elements(By.XPATH,'//*[@id]/div[1]/div/div/div[2]/span/span/span')
    retweet = driver.find_elements(By.XPATH,'//*[@id]/div[2]/div/div/div[2]/span/span/span')
    like = driver.find_elements(By.XPATH,'//*[@id]/div[3]/div/div/div[2]/span/span/span')
    print(list(map(len,[ids,comments,retweet,like])))
    for i in range(len(ids)):
        try:
            b = ids[i].text
        except:
            b=0
        try:
            c = comments[i].text
        except:
            c=0
        try:
            #print(retweet[i].text)
            d = retweet[i].text
        except:
            d=0
        try:
            e = like[i].text
        except:
            e=0
        result.append([b,c,d,e])
        driver.execute_script(f"window.scrollTo(0, {length})") 
        length += 3000
        print(result)
        time.sleep(1)
```
